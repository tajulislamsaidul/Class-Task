```python
import cv2
import matplotlib.pyplot as plt

img = cv2.imread('/content/Sample photo.jpg', cv2.IMREAD_GRAYSCALE)

v_flip = cv2.flip(img, 0)
h_flip = cv2.flip(v_flip, 1)
rotated = cv2.rotate(h_flip, cv2.ROTATE_90_CLOCKWISE)

zoomed = cv2.resize(rotated, None, fx=1.3, fy=1.3, interpolation=cv2.INTER_LINEAR)
h, w = rotated.shape
cropped_zoomed = zoomed[(zoomed.shape[0] - h) // 2:(zoomed.shape[0] - h) // 2 + h,
                         (zoomed.shape[1] - w) // 2:(zoomed.shape[1] - w) // 2 + w]


titles = ['Original', 'Vertical Flip', 'Horizontal Flip', 'Rotated', 'Zoomed']
images = [img, v_flip, h_flip, rotated, cropped_zoomed]

plt.figure(figsize=(15, 5))
for i in range(len(images)):
    plt.subplot(1, 5, i+1)
    plt.title(titles[i])
    plt.imshow(images[i], cmap='gray')
    plt.axis('off')

plt.show()
```
